```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('goodreads_train.csv')
df.head()
```

```python
import re

def list_nums(text,to_see):
    #to_see = r'\d\.\d+(?!\w)|\d\s\d\/\d(?!\w)|\d+(?!\w)'
    num_exp = re.findall(to_see, text)
    nums = []

    for i in num_exp:
        if len(re.findall(r'\d\s\d\/\d',i)) != 0:
            i = re.findall(r'\d\s\d\/\d',i)[0]
            n1 ,n2 = i.split(' ')
            n2, n3 = n2.split('/')
            nums.append(int(n1) + (int(n2)/int(n3)))
        elif len(re.findall(r'\d.\d+',i)) != 0:
            i = re.findall(r'\d.\d+',i)[0]
            nums.append(float(i))
        else:
            i = re.findall(r'\d+',i)[0]
            nums.append(float(i))
    return nums

text = df[df.review_text.str.contains('[sS]tar[s]?[\s.]?(?!\w)')].review_text

stars_indx = []
stars_list = []

ts1 = r'\d\.\d+\s[sS]tar[s]?[\s.]?(?!\w)|\d\s\d\/\d\s[sS]tar[s]?[\s.]?(?!\w)|\d+\s[sS]tar[s]?[\s.]?(?!\w)'
ts2 = r'[sS]tar[s]?:\s?\d\.\d+\s|[sS]tar[s]?:\s?\d\s\d\/\d\s|[sS]tar[s]?:\s?\d+'

to_see = [ts1,ts2]

for ts in to_see:
    for j in text.iteritems():
        ln = list_nums(j[1],ts)
        if len(ln) != 0: 
            stars_indx.append(j[0])
            stars_list.append(ln[-1])
```

### 168,000 reviews tienen stars en la review, pero solo 110,000 son las que puedo sacar

```python
ratings = list(df.loc[stars_indx].rating)

pors = 0
for i in range(len(ratings)):
    if stars_list[i] == 0.5:
        val = 1
    elif stars_list[i] == 1.5:
        val = 2
    elif stars_list[i] == 2.5:
        val = 2
    elif stars_list[i] == 3.5:
        val = 3
    elif stars_list[i] == 4.5:
        val = 4
    else:
        val = round(stars_list[i])
    
    if ratings[i] == val:
        pors += 1
print(f'Las stars en review conciden con el {pors/len(ratings)*100:0.1f}% del rating.')
```
