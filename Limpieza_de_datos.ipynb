```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('goodreads_train.csv')
df.head()
```

```python
import re
import time

def dates(date_column):
    month = {'Jan':'01','Feb':'02','Mar':'03','Apr':'04','May':'05','Jun':'06',
             'Jul':'07','Aug':'08','Sep':'09','Oct':'10','Nov':'11','Dec':'12'}

    complete_dates = []

    for i in date_column:
        if type(i) == str:
            complete_dates.append(i[-5:]+'-'+month[i[4:7]]+'-'+i[8:10]+' '+i[11:19])
        else:
            complete_dates.append(i)
    return complete_dates

to_see =['\d\.\d+\s[sS]tar[s]?[\s.]?(?!\w)|\d\s\d\/\d\s[sS]tar[s]?[\s.]?(?!\w)|\d+\s[sS]tar[s]?[\s.]?(?!\w)',
         '[sS]tar[s]?:\s?\d\.\d+\s|[sS]tar[s]?:\s?\d\s\d\/\d\s|[sS]tar[s]?:\s?\d+',
         '[rR]ating \d out of \d',
         '\(view\sspoiler\)\[',
         '\(hide\sspoiler\)\]',
         '(?<!\d)[0-5]\.?[0-9]?\s\-\s[0-5]\.?[0-9]?(?!\d)',
         '(?<!\d)[0-5]\.?[0-9]?(?!\d)\s\-',
         '\*\*\sspoiler\salert\s\*\*',
         '[0-4]\.[0-9]']
         # Agreguen estas lineas si quieren quitar el spoiler.
         #'(?<=\(view\sspoiler\)).*(?=\(hide\sspoiler\))',
         #'\(view spoiler\)\(hide spoiler\)\]',

L = len(to_see) 
print('Borrando saltos de linia (\\n)...                     ',end='\r')
df['review_text'] = df['review_text'].map(lambda x: re.sub('\n','',x))

for n,ts in enumerate(to_see):
    df['review_text'] = df['review_text'].map(lambda x: re.sub(ts,'',x))
    print(f'Tratando strellas y spoilers {n/L*100:0.2f}%              ',end='\r')
time.sleep(1)

print('Borrando espacios dobles...                        ',end='\r')
df['review_text'] = df['review_text'].map(lambda x: re.sub(r'\s\s',' ',x))

print('Reescribiendo fechas...',end='\r')
df['date_added'] = dates(df['date_added'])
df['date_updated'] = dates(df['date_updated'])
df['read_at'] = dates(df['read_at'])
df['started_at'] = dates(df['started_at'])

print('Listo                                  ')
```
